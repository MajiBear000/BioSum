```python
from summarizer import Summarizer
from tqdm import tqdm
model = Summarizer()
with open('data/test_id_jsonl.jsonl', 'w') as outfile:
    for e in tqdm(test_l):
        a = {}
        text = e['text']
        body = '. '.join(text)
        num_sentences = min(len(text), 6)
        e_sum = model(body, num_sentences=num_sentences)
        a['sent_id'] = get_sum_id(e_sum, text)
        json.dump(a, outfile)
        outfile.write('\n')
```
